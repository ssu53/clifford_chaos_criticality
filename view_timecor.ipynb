```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from tqdm.auto import tqdm

# from routines_anc import simulate_anc_pbc, simulate_anc_obc
# from routines_ent import entEntropy
```

```python
directory = '../the_cluster/anc_timecor'

reps = 10
L = 24
# ps = np.arange(0,10,1)/100
# ps = np.array([3,5,7])/100
# ps = np.arange(0,6,1)/100
ps = [0.005, 0.015, 0.025, 0.035, 0.045]
ps = np.arange(0,21,2)/100

for i in range(len(ps)):
    data_agg = np.zeros((3,0,8))
    for j in range(reps):
        infile = open(directory + '/L' + str(L) + '_p' + str(ps[i]) + '_rep' + str(j+1), 'rb')
        data = pickle.load(infile)
        infile.close()
        
        print(data.shape)
        data_agg = np.hstack((data_agg, data))
    print(data_agg.shape)
    outfile = open(directory + '/L' + str(L) + '_p' + str(ps[i]), 'wb')
    pickle.dump(data_agg, outfile)
    outfile.close()
```

```python
# go here to get colourmaps: https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
# plt.rcParams["axes.prop_cycle"] = plt.cycler("color", plt.cm.tab20c.colors)
# plt.rcParams["axes.prop_cycle"] = plt.cycler("color", plt.cm.Dark2.colors)

directory = 'anctimecor'

# L = 24
# rs = np.arange(0, L//2+3, 2)
# ps = np.arange(0,21,2)/100

# L = 48
# rs = np.arange(0,L//2,2)
# ps = np.sort(np.concatenate((np.arange(0,21,2)/100, np.array([3,5,7])/100)))

# L = 72
# rs = np.arange(0,L//2,2)
# ps = np.sort(np.concatenate((np.arange(0,10,1)/100, np.array([0.5,1.5,2.5,3.5,4.5])/100)))

L = 96
rs = np.arange(0,L//2,2)
ps = np.sort(np.concatenate((np.arange(0,6,1)/100, np.array([0.5,1.5,2.5,3.5,4.5])/100)))

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)

mis = np.empty((len(ps), len(rs)), dtype='float')
print(mis.shape)

for i in range(len(ps)):

    infile = open(directory + '/L' + str(L) + '_p' + str(ps[i]), 'rb')
    Sa, Sb, Sab = pickle.load(infile)
    infile.close()
        
    mis[i,:] = np.mean(Sa + Sb - Sab, axis=0)
    plt.plot(rs, mis[i,:], '.-')
    
plt.legend(ps)
plt.xlabel('separation(A,B)')
plt.ylabel('I(A:B)')

plt.subplot(1,2,2)
for i in range(len(rs)):
    if i % 2 == 0:
        plt.plot(ps, mis[:,i], '.-')
plt.legend(rs)
plt.xlabel('p')
plt.ylabel('I(A:B)')
```
